```python
import tldextract

def GetDomainAndSuffix(url):
    ex = tldextract.extract(url)
    return "{}.{}".format(ex.domain,ex.suffix)
```

```python
def chunks(l, n):
    """Yield successive n-sized chunks from l."""
    for i in range(0, len(l), n):
        yield l[i:i+n]
```

```python
pid_chunks = chunks(pids,1000)
chunk1 = next(pid_chunks)

elQ = "SELECT el_from, el_to FROM externallinks where el_from IN ({})".format(",".join(map(str,chunk1)))
wikiCur.execute(elQ)

els = list(wikiCur.fetchall())
el_df = pd.DataFrame(els, columns=['page_id','el_to'])
el_df['el_to'] = el_df['el_to'].str.decode('utf8')
```

```python
#takes forever, likely a shitty implementation
pres_domains = Counter()
ep_pres_domains = Counter()
samp_pres_domains = Counter()
link_counts = dict()
for k in domainCounts.keys():
    pres_domains = pres_domains + Counter(list(domainCounts[k]))
    link_counts[k] = sum(domainCounts[k].values())
    if k in ep_ids:
        ep_pres_domains = ep_pres_domains + Counter(list(domainCounts[k]))
    else:
        samp_pres_domains = samp_pres_domains + Counter(list(domainCounts[k]))
```

```python
linkStrings = dict()
for k in domainCounts.keys():
    linkStrings[k] = ' '.join([re.sub("[^A-Za-z0-9]", "", x) for x in domainCounts[k]])
```

```python
domains_for_vec = pd.DataFrame.from_dict(linkStrings,orient='index')
domains_for_vec.index.name = 'page_id'
domains_for_vec.columns = ['link_domains']
```

```python
domains_for_vec
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(analyzer="word", tokenizer=None, preprocessor=None, stop_words=None, max_features=5000)
vec_domains = vectorizer.fit_transform(domains_for_vec['link_domains'])
features = vectorizer.get_feature_names()
domain_counts = vec_domains.toarray()
domain_features = pd.DataFrame(domain_counts, columns=features, index=domains_for_vec.index)
domain_features
```

```python
total_counts = domain_features.sum(axis=0)
ep_counts = domain_features[domain_features.index.isin(ep_ids)].sum(axis=0)
samp_counts = domain_features[~domain_features.index.isin(ep_ids)].sum(axis=0)
```

```python
total_counts = total_counts / domain_features.shape[0]
ep_counts = ep_counts / len(ep_ids)
samp_counts = samp_counts / (domain_features.shape[0] - len(ep_ids))
```

```python
domains_present = pd.concat([total_counts,ep_counts,samp_counts],axis=1)
domains_present.columns = ['all','ep','sample']
domains_present.to_csv(path_or_buf="el_domains_present.csv")
```
